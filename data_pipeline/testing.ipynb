```python
import os
import shutil
```

```python
def move_images(source_root_dir, destination_dir):
  
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        print(f"Created destination directory: {destination_dir}")

    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp']  # Add more if needed

    for item in os.listdir(source_root_dir):
        item_path = os.path.join(source_root_dir, item)
        if os.path.isdir(item_path):
            print(f"Processing subdirectory: {item}")
            for filename in os.listdir(item_path):
                file_path = os.path.join(item_path, filename)
                if os.path.isfile(file_path) and any(filename.lower().endswith(ext) for ext in image_extensions):
                    destination_path = os.path.join(destination_dir, filename)
                    try:
                        shutil.move(file_path, destination_path)
                        print(f"Moved: {filename} to {destination_dir}")
                    except Exception as e:
                        print(f"Error moving {filename}: {e}")
        elif os.path.isfile(item_path) and any(item.lower().endswith(ext) for ext in image_extensions):
            # Handle images directly in the source root directory (optional)
            destination_path = os.path.join(destination_dir, item)
            try:
                shutil.move(item_path, destination_path)
                print(f"Moved (from root): {item} to {destination_dir}")
            except Exception as e:
                print(f"Error moving {item} (from root): {e}")
```

```python
source_directory = "/home/aman/code/research/CV/dia_ret/data/test/New folder/Images"
destination_directory = "/home/aman/code/research/CV/dia_ret/data/test/New folder/img"

move_images(source_directory, destination_directory)
print("Image moving process completed.")
```
